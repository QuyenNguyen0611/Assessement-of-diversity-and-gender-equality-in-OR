# Naive Bayes

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load the dataset
NB_file_path = '/content/drive/My Drive/Thesis/data.xlsx'
data_NB = pd.read_excel(NB_file_path)
```

```python
# Preprocess the data by dropping unnecessary columns
data_NB = data_NB[['name', 'gender']]

# Convert the names to strings if not already
data_NB['name'] = data_NB['name'].astype(str)

# Encode the gender labels
data_NB['gender_encoded'] = data_NB['gender'].map({'M': 0, 'F': 1})
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data_NB['name'],
    data_NB['gender_encoded'],
    test_size=0.2,
    random_state=42
)

# Vectorize the names
vectorizer = CountVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)

# Train the MultinomialNB classifier
classifier = MultinomialNB()
classifier.fit(X_train_vect, y_train)

# Predict on the test set
y_pred = classifier.predict(X_test_vect)
```

```python
# Calculate accuracy, precision, recall, f1
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
print(f'Accuracy: {accuracy * 100:.2f}%')
print(f'Precision: {precision * 100:.2f}%')
print(f'Recall: {recall * 100:.2f}%')
print(f'F1 Score: {f1 * 100:.2f}%')
```

```python
!pip install tensorflow

import pandas as pd
import numpy as np
import pickle
import json

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN, Bidirectional, LSTM, Conv1D, GlobalMaxPooling1D
from tensorflow.keras.preprocessing.sequence import pad_sequences
from imblearn.over_sampling import SMOTE
from sklearn.metrics import precision_score, recall_score, f1_score
from tensorflow.keras.metrics import Precision
from tensorflow.keras.utils import to_categorical
```

```python
# Load and prepare the dataset
# Load the dataset
file_path = '/content/drive/My Drive/Thesis/data.xlsx'
data = pd.read_excel(file_path)

data = data[['name', 'gender']]
data['name'] = data['name'].astype(str)
label_encoder = LabelEncoder()
data['gender_encoded'] = label_encoder.fit_transform(data['gender'])
```

```python
# Save the fitted LabelEncoder to a file:
with open('label_encoder.pickle', 'wb') as file:
    pickle.dump(label_encoder, file, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Tokenize the names at the character level
tokenizer = Tokenizer(char_level=True)
tokenizer.fit_on_texts(data['name'])
sequences = tokenizer.texts_to_sequences(data['name'])
max_length = max([len(seq) for seq in sequences])
vocab_size = len(tokenizer.word_index) + 1
data_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')
```

```python
# Save the fitted tokenizer to a file:
with open('tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(data_sequences, data['gender_encoded'],
                                                    test_size=0.2, random_state=42)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Function to create and compile the model
def create_compile_model(model_type, input_dim, output_dim, input_length):
    model = Sequential()
    model.add(Embedding(input_dim=input_dim, output_dim=output_dim, input_length=input_length))

    if model_type == 'RNN':
        model.add(SimpleRNN(64))
    elif model_type == 'BLSTM':
        model.add(Bidirectional(LSTM(64)))
    elif model_type == 'CNN':
        model.add(Conv1D(128, 3, activation='relu'))
        model.add(GlobalMaxPooling1D())

    model.add(Dense(64, activation='relu'))
    model.add(Dense(2, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', Precision()])
    return model
```

```python
# Initialize a dictionary to store metrics for each model type
model_metrics = {}

for model_type in ['RNN', 'BLSTM', 'CNN']:
    print(f"\nTraining and evaluating {model_type} model...")
    model = create_compile_model(model_type, vocab_size, 50, max_length)
    y_train_smote_encoded = to_categorical(y_train_smote)
    y_test_encoded = to_categorical(y_test)
    model.fit(X_train_smote, y_train_smote_encoded, batch_size=32, epochs=10, validation_data=(X_test, y_test_encoded))

    # Evaluate the model and capture all returned metrics
    evaluation_metrics = model.evaluate(X_test, y_test_encoded, verbose=0)

    # Extract specific metrics from the evaluation results
    loss = evaluation_metrics[0]
    accuracy = evaluation_metrics[1]
    precision_metric = evaluation_metrics[2]

    y_pred = model.predict(X_test, verbose=0)
    y_pred_labels = np.argmax(y_pred, axis=1)
    y_test_labels = np.argmax(y_test_encoded, axis=1)

    # Calculate metrics using scikit-learn
    precision = precision_score(y_test_labels, y_pred_labels, average='weighted')
    recall = recall_score(y_test_labels, y_pred_labels, average='weighted')
    f1 = f1_score(y_test_labels, y_pred_labels, average='weighted')

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    # Update the dictionary with metrics
    model_metrics[model_type] = {
        'loss': loss,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }
```

```python
# BLSTM is the best model, save it
model = create_compile_model('BLSTM', vocab_size, 50, max_length)
model.fit(X_train_smote, y_train_smote_encoded, batch_size=32, epochs=10, validation_data=(X_test, y_test_encoded))
model.save('model_BLSTM.h5')
```

# ANNs (BLSTMs Architecture)

```python
import re
import os
import glob
import pickle
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Load the trained model, tokenizer, and label encoder
model = load_model('model_BLSTM.h5')
with open('tokenizer.pickle', 'rb') as file:
    tokenizer = pickle.load(file)
with open('label_encoder.pickle', 'rb') as file:
    label_encoder = pickle.load(file)
```

```python
def is_abbreviated(name):
    """Check if the name is abbreviated (ends with a period)."""
    return name.endswith('.')

def extract_first_names(file_path):
    """Extracts first names from a given text file and labels abbreviated names as 'unknown'."""
    with open(file_path, 'r') as file:
        content = file.read()
    full_name_pattern = r'AUTHOR FULL NAMES: (.*?)(?=\n)'
    full_names = re.findall(full_name_pattern, content)
    names_and_labels = []
    for full_name_line in full_names:
        # Splitting names in the line
        names = [name.strip() for name in full_name_line.split(';')]
        for name in names:
            parts = name.split(',')
            if len(parts) > 1:
                first_name = parts[1].split()[0].strip()
                label = 'unknown' if is_abbreviated(first_name) else first_name
                names_and_labels.append((first_name, label))
            else:
                # Handle case where there is no comma (e.g., single name)
                label = 'unknown' if is_abbreviated(name) else name
                names_and_labels.append((name, label))
    return names_and_labels
```

```python
def preprocess_names(names):
    """Preprocess names using the tokenizer and pad them to a fixed length."""
    sequences = tokenizer.texts_to_sequences(names)
    # Adjust maxlen as per model's requirement
    padded_sequences = pad_sequences(sequences, padding='post', maxlen=15)
    return padded_sequences

def predict_gender(names):
    """Predict gender for a list of names, excluding 'unknown'."""
    preprocessed_names = preprocess_names(names)
    predictions = model.predict(preprocessed_names)
    predicted_labels = np.argmax(predictions, axis=1)
    return label_encoder.inverse_transform(predicted_labels)

def extract_names_from_files(folder_path):
    """Extracts first names and their labels from all text files in the given folder."""
    file_paths = glob.glob(os.path.join(folder_path, '*.txt'))
    results = {}
    for path in file_paths:
        file_name = os.path.basename(path)
        results[file_name] = extract_first_names(path)
    return results

# Specify the folder path
folder_path = '/content/drive/My Drive/Thesis/TXT file'

# Extract names and labels
extracted_names_and_labels = extract_names_from_files(folder_path)
```

```python
# Mapping model output to human-readable labels
gender_map = {'M': 'male', 'F': 'female'}

# Applying the model to predict gender for each name in each file
for file_name, names_and_labels in extracted_names_and_labels.items():
    # Reset gender counts for each file
    gender_counts = {'male': 0, 'female': 0, 'unknown': 0}

    names_to_predict = [name for name, label in names_and_labels if label != 'unknown']

    if names_to_predict:
        predicted_genders = predict_gender(names_to_predict)
    else:
        predicted_genders = []

    print(f"Predicted genders for names in {file_name}:")
    predicted_gender_index = 0

    # Process each name and label
    for name, label in names_and_labels:
        if label == 'unknown':
            print(f"{name}: Unknown")
            gender_counts['unknown'] += 1
        else:
            model_output = predicted_genders[predicted_gender_index]
            # Use 'unknown' for any unexpected output
            gender = gender_map.get(model_output, 'unknown')
            #print(f"{name}: {gender.capitalize()}")
            gender_counts[gender] += 1
            predicted_gender_index += 1

    # Print the gender counts for the current file
    print("\nGender Counts for", file_name)
    for gender, count in gender_counts.items():
        print(f"{gender.capitalize()}: {count}")
    print("\n")  # Adds an extra newline for better readability between files
```
